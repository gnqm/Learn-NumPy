```python
# NumPy (Numerical Python)
import numpy as np

a=[1,2,3,4,5]
print("This is a list:",a)
b=np.array(a)
print("This is an array:",b)
print("Class of array:",type(b)) # not a list
print("Datatype of array:",b.dtype) # dtype attribute returns the datatype of elements in array
print("Appending elements:",np.append(a,[6,7]))

c=[[1,2,3],[4,5,6],[7,8,9]]
d=np.array(c)
print("\n","Two-dimensional array (matrix):\n",d)
print("Second Element: ",d[1])
print("Extracting a sub-matrix using indexing:\n",d[:2,1:])

print("\nThis is a list",list(range(0,5))," using range()")
a=np.arange(0,5) # works exactly similar to range()
print("This is an array",a," using arange()\n")

# creating an array of zeros or ones or any other number
print("One dimensional zeros: ",np.zeros(5))
print("\nTwo dimensional ones:(rows x columns)\n",np.ones((2,3)))
print("\nTwo dimensional fives: (3x5)\n",(5*np.ones((3,5))))
```

```python
import numpy as np

# creating linearly spaced numbers using linspace()
print("\nLinearly spaced numbers: ",np.linspace(0,10,6)) # the first and last numbers are included along with the equally spaced remaining numbers
print("A number line having points from [-5,5]: ",np.linspace(-5,5,11))

# creating an identity matrix
print("\nAn identity matrix is defined using eye():\n",np.eye(5),"\n")

# generating a random number using uniform distribution (probability of each number between 0 and 1 is the same)
print("One-dimensional random numbers using rand(): ",np.random.rand(4),"\n")
print("Two-dimensional random numbers using rand(): \n",np.random.rand(3,5))

# generating a random number using normal distribution (Gaussian distribution with mean=0 and variance=1)
print("\nOne-dimensional random numbers using randn(): ",np.random.randn(4),"\n")
print("Two-dimensional random numbers using randn(): \n",np.random.randn(3,5))

# generating random numbers between low and High values
a=np.random.randint(1,100,20)
print("\n20 numbers generated randomly between [1,100) are: ",a)
print("\nShape:",a.shape)# shape attribute returns the shape
print("Size:",a.size)# size attribute returns the number of elements
print("Dimension:",a.ndim)# ndim attribute returns the dimension or axes
b=a.reshape(4,5) # reshape() is used to change the dimension
c=a.reshape(2,10)
print("\nThe above array can be re-arranged into 4x5 using reshape():\n",b)
print("\nShape:",b.shape)
print("Size:",b.size)
print("Dimension:",b.ndim)
print("\nThe above array can be re-arranged into 2x10 using reshape():\n",c)
print("\nShape:",c.shape)
print("Size:",c.size)
print("Dimension:",c.ndim)

# max and argmax()
print("\nArray: ",a)
print("Returning maximum number using max(): ",a.max())
print("Returning minimum number using min(): ",a.min())
print("Returning index position of maximum number using argmax(): ",a.argmax())
print("Returning index position of minimum number using argmin(): ",a.argmin())
```

```python
# NumPy Operations
import numpy as np

a=np.random.randint(1,100,10)
print("\nOriginal Array (a): ",a)
print("Addition (a+a): ",a+a)
print("Subtraction (a-10): ",a-10)
print("Multiplication (a*2): ",a*2)
print("Division (a/5): ",a/5)
print("Exponent (a**2): ",a**2)

# Other functions
pi=np.pi
print("\nValue of pi: ",pi)
print("Ceil Value of pi: ",np.ceil(pi)) # ceil() is used to approximate the value on the higher side
print("Floor Value of pi: ",np.floor(pi)) # floor() is used to approximate the value on the lower side
print("Around Value of pi: ",np.around(pi,decimals=4)) # around() is used to round the result into specified decimals places. 

a=np.random.randint(1,100,5)
print("\nOriginal Array: ",a)
print("Square Root: ",np.around(np.sqrt(a),decimals=2))# sqrt() is used to determine the square root of array
print("Exponential: ",np.exp(a))# Exponential
print("Power: ",np.power(a,2)) # power is used to find the power of an array.
print("Maximum: ",np.max(a))
print("Minimum: ",np.min(a))
print("Sinusoid: ",np.around(np.sin(a),decimals=3))
print("Logarithm: ",np.around(np.log(a),decimals=3))
print("Mean: ",np.mean(a))
print("Standard Deviation: %.2f"%(np.std(a)))
print("Variance: ",np.var(a))
print("Sum: ",np.sum(a))
print("Median: ",np.median(a))
```

```python
# Conditional Selection
import numpy as np

# Method_1
a=np.arange(1,11)
print("Array:",a)
b=a>5
print("Result of Array:",b) # returns only true values
print("Filtered Array:",a[b]) # returns true values in the array

# Method_2
print("Filtered Array:",a[a>5]) # combining above two conditions in a single statement
```

```python
import numpy as np

a=np.arange(1,7).reshape(2,3)
print("Matrix: (2x3)\n",a)
print("Sum of elements (axis=0): \n",a.sum(axis=0)) # axis=0 means columns=fixed, rows=variable
print("Sum of elements (axis=1): \n",a.sum(axis=1)) # axis=1 is means columns=variable, rows=fixed
print("Sum of all elements: \n",a.sum())
print("Matrix after transformation: (3x2)\n",a.T) # T attribute is used to transform the matrix
print("Appending elements along axis 0:\n",np.append(a,[[7,8,9],[10,11,12]],axis=0))
print("Appending elements along axis 1 horizontally:\n",np.append(a,[[7,8],[9,10]],axis=1))
print("Appending elements along axis 1 vertically:\n",np.append(a,[[7],[8]],axis=1))
print("\nMaximum along each column: ",np.amax(a,1)) # amax(array,axis) is used to determine the max. along the specified axis
print("Minimum along each row: ",np.amin(a,0)) # amin(array,axis) is used to determine the min. along the specified axis
print("Range along each column: ",np.ptp(a,1)) # ptp(array,axis) is used to determine the range along the specified axis
print("Percentile along each row: ",np.percentile(a,50,0)) # percentile(array,percentile,axis) is used to determine the percentile along the specified axis
print("Overall Percentile: ",np.percentile(a,40)) 
print("Median along each column: ",np.median(a,1)) # median(array,axis) is used to determine the median along the specified axis
print("Mean along each row: ",np.mean(a,0)) # mean(array,axis) is used to determine the mean along the specified axis
a=np.array([12,5])
print("\na:",a)
print("Pythagoras Value using np.linalg.norm(a):",np.linalg.norm(a)) # norm() is basically the pythagoras value
b=np.array([3,4])
print("b:",b)
print("Pythagoras Value using np.linalg.norm(b):",np.linalg.norm(b))
```

```python
import numpy as np
x=np.array([[1,3],[4,2]])
y=np.array([[2,6],[2,7]])
print("x is:\n",x)
print("y is:\n",y)
print("\nDeterminant of x: ",np.linalg.det(x))
print("Inverse of y:\n",np.linalg.inv(y))
print("Diagonal of x:",np.diag(x)) # np.diag(matrix) provides us with the array of diagonal elements
print("Sum of Diagonal elements of y:",np.trace(y)) # np.trace(matrix) provides us with the sum of diagonal elements
print("Creating Diagonal matrix:\n",np.diag(np.diag(y))) # np.diag(array/list) will provide us the diagonal matrix with other elements as 0

# Element-wise operations
print("\nx+y is:\n",x+y)
print("x-y is:\n",x-y)
print("x*y is:\n",x*y) # element-wise product
print("x.y is:\n",np.dot(x,y)) # matrix multiplication using dot()
print("\nDot Product of two vectors: ",np.vdot(x,y)) # vdot() is used for dot product of vectors, adding individual elements
print("Inner product is similar to the dot product at each position of the matrix:\n",np.inner(x,y)) # inner() is same as vdot(), row to row multiplication
print("Outer product:\n",np.outer(x,y)) # outer product is the mapping of y on x, size of the result is (No. of elements in x * No. of elements in y)
print("\nConcatenating x & y along axis 0:\n",np.concatenate((x,y))) # concatenating along axis=0 (default)
print("Concatenating x & y along axis 1:\n",np.concatenate((x,y),axis=1)) # concatenating along axis=1
print("Horizontal Stacking of x & y:\n",np.hstack((x,y)))
print("Vertical Stacking of x & y:\n",np.vstack((x,y)))
print("Sort along each row:\n",np.sort(np.vstack((x,y)),0)) # sort(array,axis) is used to sort elements along the specified axis
```

```python
# Solving Linear System (Ax=b) -> (x=(inverse of A).b)
import numpy as np

A=np.floor(np.random.random((2,2))*20)
print("a is:\n",A)
b=np.ceil(np.random.random(2)*10)
print("b is:\n",b)

# Method 1
a_inv=np.linalg.inv(A)
print("a_inverse is:\n",a_inv)
x=np.dot(a_inv,b)
print("Result is: ",x)

# Method 2
x=np.linalg.solve(A,b) # solve() is used to find the solution of the linear system
print("Result using solve() is: ",x)
```

```python
# Example a.b=|a||b|cos(theta)
import numpy as np

# Generating two matrices 'a' and 'b'
a=np.around((np.random.randint(-10,10,9)-5))
print("Matrix a:\n",a.reshape(3,3))
b=np.ceil(np.random.randn(3,3)*10)
print("\nMatrix b:\n",b)

np.rank
dp=np.vdot(a,b) # dot product
print("\nDot Product of a and b: ",dp)
x=np.linalg.norm(a)
y=np.linalg.norm(b)
print("Abs Values of a is {} and b is {}: ".format(x,y))
costheta=dp/(x*y)
angle=np.rad2deg(np.arccos(costheta))
print("Angle is: ",np.ceil(angle)," degrees")
```
